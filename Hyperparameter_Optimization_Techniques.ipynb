I'd be glad to provide practical examples of hyperparameter optimization methods using GridSearchCV, RandomizedSearchCV, and BayesianSearchCV:

**1. GridSearchCV**

**Scenario:** You're building a support vector machine (SVM) for sentiment analysis and want to optimize the `C` (regularization) and `kernel` hyperparameters.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}

# Create the SVM model
svc = SVC()

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5)  # 5-fold cross-validation

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)

# Best parameters and score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Print results
print("Best parameters:", best_params)
print("Best cross-validation score:", best_score)

# Use the best model on the test data
best_model = grid_search.best_estimator_
predictions = best_model.predict(X_test)
```

**Explanation:**

- `param_grid` defines the ranges for each hyperparameter to be explored.
- `GridSearchCV` creates a grid of all possible hyperparameter combinations and evaluates each combination using cross-validation.
- `.fit()` trains models with each combination on training folds and evaluates them on validation folds.
- `best_params_` and `best_score_` store the best hyperparameter combination and its corresponding score.
- You can then use the `best_estimator_` to make predictions on unseen data.

**2. RandomizedSearchCV**

**Scenario:** You have a large hyperparameter space and want to be more efficient than GridSearchCV. RandomizedSearchCV samples a random subset of hyperparameter combinations.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC

# Define the parameter distribution (can be more flexible than grid)
param_dist = {
    'C': np.random.uniform(0.01, 100),
    'kernel': ['linear', 'rbf']
}

# Create the SVM model
svc = SVC()

# Set up RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=svc, param_distributions=param_dist, cv=5, n_iter=100)  # 100 random iterations

# Fit the randomized search to the training data
random_search.fit(X_train, y_train)

# Best parameters and score
best_params = random_search.best_params_
best_score = random_search.best_score_

# Print results
print("Best parameters:", best_params)
print("Best cross-validation score:", best_score)

# Use the best model on the test data
best_model = random_search.best_estimator_
predictions = best_model.predict(X_test)
```

**Explanation:**

- `param_dist` uses probability distributions for hyperparameters (e.g., uniform distribution for `C`).
- `RandomizedSearchCV` randomly samples combinations from these distributions and evaluates them.
- `n_iter` controls the number of random iterations to perform.
- Similar to GridSearchCV, the best parameters and score are stored, and you can use the `best_estimator_` for predictions.

**3. BayesianSearchCV**

**Scenario:** You want a more adaptive approach that learns from the results of previous evaluations to focus on promising hyperparameter regions.

```python
from sklearn.model_selection import BayesianSearchCV
from sklearn.svm import SVC

# Define the parameter space (can be more complex than a grid)
param_space = {
    'C': (0.01, 100, 'uniform'),
    'kernel': ['linear', 'rbf']
}

# Create the SVM model
svc = SVC()

# Set up BayesianSearchCV
bayesian_search = BayesianSearchCV(estimator=svc, param_space=param_space, cv=5)

# Fit the Bayesian search to the training data
bayesian_search.fit(X_train, y_train)

# Best parameters and score
best_params = bayesian_search.best_params_
best_score = bayesian_search
```

### 4. **Hyperband**
   - **Description:** Extends Random Search with early stopping, allowing it to allocate more resources to promising hyperparameters and fewer to those that perform poorly.
   - **Pros:** Faster than Random Search and Grid Search; reduces computational cost by terminating poor configurations early.
   - **Cons:** Works best with iterative algorithms that can be interrupted, such as neural networks.

### 5. **Genetic Algorithms**
   - **Description:** Uses evolutionary strategies, including selection, mutation, and crossover, to evolve a population of hyperparameter configurations over successive generations.
   - **Pros:** Good for large and complex search spaces; can escape local optima.
   - **Cons:** Computationally intensive; may require many generations to converge to an optimal solution.

### 6. **Sequential Model-Based Optimization (SMBO)**
   - **Description:** Similar to Bayesian Optimization but typically uses different probabilistic models. The algorithm sequentially selects new hyperparameter configurations based on past evaluations.
   - **Pros:** Efficient and suitable for high-dimensional spaces; adaptively learns and improves search efficiency.
   - **Cons:** Complex and can be slower than simpler methods for smaller datasets.

### 7. **Grid Search with Cross-Validation**
   - **Description:** Combines Grid Search with cross-validation to evaluate each hyperparameter configuration's performance more robustly.
   - **Pros:** Provides more reliable estimates of performance; useful for smaller datasets.
   - **Cons:** Computationally expensive, especially with large grids and datasets.
